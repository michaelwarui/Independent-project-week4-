```
#importing the libraries that we will require
import pandas as pd
import numpy as np   
```

```
#creating a dataframe from the data
#loading our data from the csv file 
Auto=pd.read_csv('Autolib_dataset .csv')
Auto
```

```
Auto['Bluecar usage']=Auto['Bluecar counter'].diff()
Auto['Utilib usage']=Auto['Utilib counter'].diff()
Auto['Utilib 1.4 usage']=Auto['Utilib 1.4 counter'].diff()
```

```
Auto
```

```
#Accesing information from our dataset
print(Auto.info())
```

```
#cheaking if any null values exist
#
print(Auto.isnull().sum())
```

```
#dropping the displayed comment column in the data
#As it contains to many missing values 
Auto.drop(['Displayed comment'], axis= 1,inplace=True)
Auto
```

```
Auto.head(3)
```

```
#combining the separate year, month, day , hour and minute column into one datetime column
Auto['Int']=Auto['year'].astype(str) + Auto['month'].astype(str).str.zfill(2)+ Auto['day'].astype(str).str.zfill(2) + Auto['hour'].astype(str).str.zfill(2) + Auto['minute'].astype(str).str.zfill(2)
Auto['Datetime'] = pd.to_datetime(Auto['Int'], format='%Y%m%d%H%M')
Auto

```

```
#dropping the year, month, day, hour and minute column so as to remain with the combinded Datetime column
Auto.drop(['year','month','day','hour','minute','Int'], axis=1)
Auto
```

```
#arranging the data into ascending order
Auto.sort_values(by='Datetime', ascending=1,inplace=True)
Auto
```

```
#creating a usage column for each of the cars
Auto['Bluecar_usage']=Auto['Bluecar counter'].diff()
Auto['Utilib_usage']=Auto['Utilib counter'].diff()
Auto['Utilib1.4_usage']=Auto['Utilib 1.4 counter'].diff()
```

```
#previewing the updated column
Auto
```

```
#where the usage result is negative the car was taken.
#where the usage result is positive the car was returned.
#where the usage is 0.0 this means that nothing was taken or returned.
Auto[['Bluecar counter',	'Utilib counter',	'Utilib 1.4 counter','Bluecar_usage',	'Utilib_usage',	'Utilib1.4_usage']]
```
